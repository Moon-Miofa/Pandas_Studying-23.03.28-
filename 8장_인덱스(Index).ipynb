#8. 인덱스(Index)

인덱스 : 데이터에 접근할 수 있는 기준 또는 주소 또는 구분값

-> 유일값

```python
data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고','북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학':[100,50,70,70,10,95,45,90],
    '과학':[95,55,80,75,35,85,40,95],
    '사회':[85,25,75,80,10,80,35,95],
    'sw특기':['Python','Java','Javascript','','','C','PYTHON','C#']
}
```

```python
import pandas as pd
```

```python
pd.DataFrame(data)
```

```python
df = pd.DataFrame(data) #위의 데이터 프레임 할당
```

```python
df.index
```

```python
df.columns
```

```python
df = pd.DataFrame(data, index= ['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번'])
#인덱스 지정
```

```python
df
```

```python
df.index
```

##인덱스 이름 지정

```python
df.index.name = '지원번호'
#인덱스의 이름을 지정. **열 이름과는 다르다.
```

```python
df
```

##인덱스 초기화

```python
df.reset_index()
#인덱스 지우기
#기존의 인덱스가 '열'로 포함된다.
```

```python
df
#재실행하면, 다시 인덱스가 살아있다??
#reset_index()는 인덱스가 사라진 모습을 보여주고, 사라진다 -> 저장되는 것이 아님.
```

```python
df.reset_index(drop=True)
```

```python
"""
방법1. 언제나처럼 할당을 해준다.
df = df.reset_index(drop=True)
"""
```

```python
df.reset_index(drop=True, inplace=True)
#방법2. inplace = True 구분을 사용.
```

```python
df
```

##지정한 column으로 인덱스 설정

```python
df.set_index('이름')
#'이름' column을 인덱스로 설정 *이 경우 '이름'데이터에는 중복이 없다고 생각해야 한다.
```

```python
df.set_index('이름', inplace=True)
```

```python
df
```

##인덱스 정렬

```python
df.sort_index()
```

```python
df.sort_index(inplace=True)
#sort_index()는 오름차순순
```

```python
df
```

```python
df.sort_index(ascending=False, inplace=True)
#기본적으로 asceding 값은 True, 직접 False로 바꿔주면 내림차순 정렬이 된다
```
