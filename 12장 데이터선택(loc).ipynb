#12. 데이터선택(loc)

loc : location, 행 index, 열 index 이름을 이용해서 데이터를 선택할 때

```python
import pandas as pd
```

```python
df = pd.read_excel(drive_path + 'score.xlsx',index_col = '지원번호')
```

```python
df.loc['1번']
```

```python
df.loc['5번']
```

```python
df.loc['5번']['이름':'키']
```

```python
#행 index와 열 index의 만나는 값을 간단하게 불러올 수 있다.
df.loc['1번','이름']
```

```python
df.loc[['1번', '5번', '7번']]
```

```python
df.loc[['1번', '5번', '7번'],['이름', '학교', '수학']]
```

```python
df.loc['2번':'5번']
```

```python
#열과 행에 대해 모두 슬라이싱이 가능하다.
df.loc['2번':'5번', '국어':'사회']
```

→loc를 사용하지않은 슬라이싱 : 범위가 인덱스

→loc를 사용한 슬라이싱 : 명시된 인덱스를 보이는 그대로


```python
#원본이 아니라 슬라이싱을 통한 복제본이 나왔다.
df.loc[:]
```

```python
df.loc[:, :]
```

```python
df.loc[:, :'사회']
```

```python
#loc는 명시적인 '이름'으로 설정할 수 있다. // 숫자는 불가능
df.loc[:, :-2]
```
