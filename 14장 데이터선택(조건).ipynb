#논리값(참, 거짓)을 이용해서 데이터 선택

```python
#2번, 3번은 보이지 않는다. → False의 값은 가져오지 않는다.
df.loc[[True, False, False, True, True, True, True, True]]
```

```python
#조건 참조가 가능
df.loc[df['키'] >= 185]
```

```python
#대괄호로 표시하면 1개의 행에 대해서도 데이터프레임으로 가져온다.
df.loc[df['키'] >=185, ['이름']]
```

```python
df.loc[df['키'] >= 185, ['이름', '키', '학교']]
```

```python
df.loc[df['키'] >= 185, '국어':'사회']
```

##조건 연결 : &, and 조건, 단일 조건들이 모두 맞아야 전체가 참인 경우

```python
(df['키'] >= 185) & (df['키'] < 200)
```

```python
#괄호로 단일 조건을 나누지 않으면 에러
df.loc[df['키'] >= 185 & df['키'] < 200]
```

```python
df.loc[(df['키']>=185) & (df['키'] < 200)]
```

```python
df.loc[(df['키']>=185) & (df['학교'] == '북산고')]
```

##조건 연결 : | or조건, 단일 조건 중 하나라도 참이면 연결 조건 전체가 참인 경우

```python
#키가 175보다 작다 (또는, or) 키가 200보다 큰 경우
(df['키']  <175) | (df['키'] > 200)
```
