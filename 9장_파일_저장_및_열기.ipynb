CSV파일을 엑셀에서 연다면 글시가 깨진다.

텍스트 표현법 : 인코딩 방식 / 코드 방식

ANSI, UTF-8 <- (온라인 표준 / 맥, 리눅스 채택)

우리말 : euc-kr, cp949 (윈도우 기본)

기본 저장이 utf-8 인코딩이기에 엑셀에서는 깨져보인다

윈도우 기준인 cp949로 저장

```python
#encoding = 텍스트 인코딩 방식식
df.to_csv(drive_path + 'score.csv', encoding='cp949')
```

모든 운영체제와 온라인 환경에서 텍스트 표현이 가능한 인코딩 방식 : utf-8-sig

```python
df.to_csv(drive_path + 'score2.csv', encoding='utf-8-sig')
```

```python
#인덱스를 제외하고 저장하기
df.to_csv(drive_path + 'score.csv', encoding='utf-8-sig', index=False)
```
