#13. 데이터선택(iloc)

iloc : index location (암묵적인 위치 주소로 데이터를 선택하는 방법)

    integer location으로도 불림

```python
from google.colab import drive
drive.mount('/gdrive')
```

```python
import pandas as pd
```

```python
df.iloc[[0, 3, 5]]
```

```python
df.iloc[:4]
```

```python
df.iloc[:4, 2:]
```

```python
df.iloc[:, -1]
```

```python
df.iloc[:, :-1]
```
